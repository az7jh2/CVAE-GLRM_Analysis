# Run *SDePER* on Real data: MOB

In this Notebook we run SDePER on one Real Dataset -- mouse olfactory bulb (**MOB**). 

For downloading and preprocessing original spatial and reference scRNA-seq data for cell type deconvolution please refer [MOB_preprocess.nb.html](https://rawcdn.githack.com/az7jh2/SDePER_Analysis/0a429908645a665c1f9d345df013d5b9fcde20b3/RealData/MOB/MOB_preprocess.nb.html).

==================================================================================================================

So here we use **5 input files** as shown below:

1. raw nUMI counts of spatial transcriptomic data (spots × genes): [MOB_spatial_spot_nUMI.csv](https://github.com/az7jh2/SDePER_Analysis/blob/main/RealData/MOB/MOB_spatial_spot_nUMI.csv)
2. raw nUMI counts of reference scRNA-seq data (cells × genes): `MOB_ref_scRNA_cell_nUMI.csv`, just decompress the gzipped file [MOB_ref_scRNA_cell_nUMI.csv.gz](https://github.com/az7jh2/SDePER_Analysis/blob/main/RealData/MOB/MOB_ref_scRNA_cell_nUMI.csv.gz).
3. cell type annotations for cells of selected 5 cell types in reference scRNA-seq data (cells × 1): [MOB_ref_scRNA_cell_celltype.csv](https://github.com/az7jh2/SDePER_Analysis/blob/main/RealData/MOB/MOB_ref_scRNA_cell_celltype.csv)
4. adjacency matrix of spots in spatial transcriptomic data (spots × spots): [MOB_spatial_spot_adjacency_matrix.csv](https://github.com/az7jh2/SDePER_Analysis/blob/main/RealData/MOB/MOB_spatial_spot_adjacency_matrix.csv)
5. row/column integer index (x,y) of spatial spots (spots * 2): [MOB_spatial_spot_loc.csv](https://github.com/az7jh2/SDePER_Analysis/blob/main/RealData/MOB/MOB_spatial_spot_loc.csv)

==================================================================================================================

SDePER settings are:

* number of included highly variable genes `n_hv_gene`: 250
* number of selected TOP marker genes for each comparison in Differential `n_marker_per_cmp`: 20
* minimum value of cells in one pseudo-spot for building CVAE `pseudo_spot_min_cell`: 10
* maximum value of cells in one pseudo-spot for building CVAE `pseudo_spot_max_cell`: 40
* initial learning rate for training CVAE `cvae_init_lr`: 0.001
* seed for random values `seed`: 13
* number of used CPU cores `n_core`: 20
* whether to perform imputation `use_imputation`: true

ALL other options are left as default.

==================================================================================================================

the `bash` command to start cell type deconvolution is

`runDeconvolution -q MOB_spatial_spot_nUMI.csv -r MOB_ref_scRNA_cell_nUMI.csv -c MOB_ref_scRNA_cell_celltype.csv -a MOB_spatial_spot_adjacency_matrix.csv -l MOB_spatial_spot_loc.csv --n_hv_gene 250 --n_marker_per_cmp 20 --pseudo_spot_min_cell 10 --pseudo_spot_max_cell 40 --cvae_init_lr 0.001 --seed 13 -n 20 --use_imputation true`

Note this Notebook uses **SDePER v1.0.1**. Cell type deconvolution result is renamed as [MOB_SDePER_celltype_proportions.csv](https://github.com/az7jh2/SDePER_Analysis/tree/main/RealData/MOB/MOB_SDePER_celltype_proportions.csv). Imputation results are compressed into one zipped file and deposited to folder [MOB](https://zenodo.org/record/8334656/files/MOB.zip) in [10.5281/zenodo.8334655](https://doi.org/10.5281/zenodo.8334655).

```python
import subprocess

cmd = '''runDeconvolution -q MOB_spatial_spot_nUMI.csv \
                          -r MOB_ref_scRNA_cell_nUMI.csv \
                          -c MOB_ref_scRNA_cell_celltype.csv \
                          -a MOB_spatial_spot_adjacency_matrix.csv \
                          -l MOB_spatial_spot_loc.csv \
                          --n_hv_gene 250 \
                          --n_marker_per_cmp 20 \
                          --pseudo_spot_min_cell 10 \
                          --pseudo_spot_max_cell 40 \
                          --cvae_init_lr 0.001 \
                          --seed 13 \
                          -n 20 \
                          --use_imputation true
'''
subprocess.run(cmd, check=True, text=True, shell=True)
```
